```python
from ultralytics import YOLO
import os
from PIL import Image

def get_image_paths_from_folder(folder_path):
    """
    从指定的文件夹中递归获取所有图像文件的路径。

    :param folder_path: 包含图像的文件夹路径。
    :return: 图像文件路径的列表。
    """
    image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')  # 支持的图像格式
    image_paths = []
    for root, _, files in os.walk(folder_path):  # 递归遍历文件夹
        for file in files:
            if file.lower().endswith(image_extensions):  # 检查文件扩展名
                image_paths.append(os.path.join(root, file))
    return image_paths

def process_images_with_background(image_paths, output_root, model):
    """
    处理一组图像，将检测到的目标（带背景）裁剪并保存到以图片名称命名的子文件夹中。

    :param image_paths: 包含所有待处理图像路径的列表。
    :param output_root: 指定的结果输出根目录。
    :param model: 已加载的 YOLO 模型实例。
    """
    # 创建主输出文件夹 Seg_with_Bg
    output_root = os.path.join(output_root, "Seg_with_Bg")
    os.makedirs(output_root, exist_ok=True)

    for image_path in image_paths:
        try:
            # 加载原始图像
            original_image = Image.open(image_path)

            # 获取原始图像的文件名（不带后缀）并创建子文件夹
            base_name = os.path.splitext(os.path.basename(image_path))[0]
            output_folder = os.path.join(output_root, base_name)
            os.makedirs(output_folder, exist_ok=True)  # 如果文件夹不存在，则创建

            # 使用 YOLO 模型进行目标检测
            results = model(image_path)

            # 检查是否有检测结果
            if len(results[0].boxes) == 0:
                print(f"图像 {image_path} 没有检测到目标，跳过...")
                continue

            # 遍历检测结果中的每个目标
            for i, box in enumerate(results[0].boxes.xyxy):  # 获取每个目标的边界框 (x1, y1, x2, y2)
                x1, y1, x2, y2 = map(int, box)  # 将坐标转换为整数
                cropped_image = original_image.crop((x1, y1, x2, y2))  # 裁剪图像

                # 构造保存文件的路径
                output_path = os.path.join(output_folder, f"{base_name}_{i + 1}.jpg")

                # 保存裁剪后的图像
                cropped_image.save(output_path)
                print(f"目标 {i + 1} 已保存到: {output_path}")

        except Exception as e:
            print(f"处理图像 {image_path} 时出错，跳过该图像。错误信息: {e}")
            continue

    print(f"所有图像处理完成，结果已保存到: {output_root}")

# 示例用法
input_folder = "C:/Users/14122/Desktop/test data/Met-Meme/test_data_En_ver"  # 包含图像的文件夹
output_root = "C:/Users/14122/Desktop/test data/Met-Meme/YOLO_Object_Detection/Results"

# 加载 YOLO 模型
model = YOLO("yolo11n.pt")  # 确保模型权重文件路径正确

# 获取文件夹中的所有图像路径
image_paths = get_image_paths_from_folder(input_folder)

# 处理图像
process_images_with_background(image_paths, output_root, model)
```

```python
from ultralytics import YOLO
import os
import cv2
import numpy as np

def get_image_paths_from_folder(folder_path):
    """
    从指定的文件夹中递归获取所有图像文件的路径。

    :param folder_path: 包含图像的文件夹路径。
    :return: 图像文件路径的列表。
    """
    image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')  # 支持的图像格式
    image_paths = []
    for root, _, files in os.walk(folder_path):  # 递归遍历文件夹
        for file in files:
            if file.lower().endswith(image_extensions):  # 检查文件扩展名
                image_paths.append(os.path.join(root, file))
    return image_paths

def log_error(log_file, message):
    """
    将错误信息写入日志文件。

    :param log_file: 日志文件路径。
    :param message: 要记录的错误信息。
    """
    with open(log_file, 'a') as f:
        f.write(message + '\n')

def process_images_without_background(image_paths, output_root, model, log_file):
    """
    处理一组图像，将检测到的目标（去掉背景）裁剪并保存到以图片名称命名的子文件夹中。

    :param image_paths: 包含所有待处理图像路径的列表。
    :param output_root: 指定的结果输出根目录。
    :param model: 已加载的 YOLO 模型实例。
    :param log_file: 日志文件路径。
    """
    # 创建主输出文件夹 Seg_without_Bg
    output_root = os.path.join(output_root, "Seg_without_Bg")
    os.makedirs(output_root, exist_ok=True)

    for image_path in image_paths:
        try:
            # 检查路径是否存在
            if not os.path.exists(image_path):
                message = f"图像路径不存在，跳过: {image_path}"
                print(message)
                log_error(log_file, message)
                continue

            # 加载原始图像
            original_image = cv2.imread(image_path)
            if original_image is None:
                message = f"无法读取图像，跳过: {image_path}"
                print(message)
                log_error(log_file, message)
                continue

            # 获取原始图像的文件名（不带后缀）并创建子文件夹
            base_name = os.path.splitext(os.path.basename(image_path))[0]
            output_folder = os.path.join(output_root, base_name)
            os.makedirs(output_folder, exist_ok=True)  # 如果文件夹不存在，则创建

            # 使用 YOLO 模型进行目标检测
            results = model(image_path)

            # 检查是否有检测结果
            if results[0].masks is None or len(results[0].masks.data) == 0:
                message = f"图像 {image_path} 没有检测到目标，跳过..."
                print(message)
                log_error(log_file, message)
                continue

            # 遍历检测结果中的每个目标
            for i, mask in enumerate(results[0].masks.data):  # 获取每个目标的掩码
                # 将掩码转换为二值图像
                mask = mask.cpu().numpy().astype(np.uint8) * 255

                # 调整掩码大小以匹配原始图像
                mask_resized = cv2.resize(mask, (original_image.shape[1], original_image.shape[0]), interpolation=cv2.INTER_NEAREST)

                # 提取目标区域
                target = cv2.bitwise_and(original_image, original_image, mask=mask_resized)

                # 获取掩码的边界框
                x, y, w, h = cv2.boundingRect(mask_resized)

                # 裁剪目标区域
                cropped_image = target[y:y+h, x:x+w]

                # 构造保存文件的路径
                output_path = os.path.join(output_folder, f"{base_name}_{i + 1}.png")

                # 保存裁剪后的图像
                cv2.imwrite(output_path, cropped_image)
                print(f"目标 {i + 1} 已保存到: {output_path}")

        except Exception as e:
            message = f"处理图像 {image_path} 时出错，跳过该图像。错误信息: {e}"
            print(message)
            log_error(log_file, message)
            continue

    print(f"所有图像处理完成，结果已保存到: {output_root}")

# 主函数
if __name__ == "__main__":
    # 输入文件夹路径
    input_folder = "C:/Users/14122/Desktop/test data/Met-Meme/test_data_En_ver/1"  # 包含图像的文件夹
    output_root = "C:/Users/14122/Desktop/test data/Met-Meme/YOLO_Object_Detection/Results"
    log_file = "C:/Users/14122/Desktop/test data/Met-Meme/YOLO_Object_Detection/error_log.txt"  # 日志文件路径

    # 加载 YOLO 模型
    model = YOLO("yolo11n-seg.pt")  # 确保模型权重文件路径正确

    # 获取文件夹中的所有图像路径
    image_paths = get_image_paths_from_folder(input_folder)

    # 处理图像
    process_images_without_background(image_paths, output_root, model, log_file)
```
