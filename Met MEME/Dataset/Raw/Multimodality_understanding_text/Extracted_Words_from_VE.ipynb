```python
%pip install spacy
import spacy

# Download the model if not already present
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    from spacy.cli import download
    download("en_core_web_sm")
    nlp = spacy.load("en_core_web_sm")

text = "The image shows a sad-looking seal with tears in its eyes, paired with the caption \"I'M FUCKIN LIT.\""
doc = nlp(text)

# Extract nouns and noun phrases
keywords = set()
for chunk in doc.noun_chunks:
    keywords.add(chunk.root.text)
for token in doc:
    if token.pos_ == "NOUN":
        keywords.add(token.text)

print(list(keywords))
```

```python
d_a = pd.read_csv(r"F:\Github\MemePic\MemePic\Met MEME\Dataset\Raw\Multimodality_understanding_text\d_a.csv")
d_b = pd.read_csv(r"F:\Github\MemePic\MemePic\Met MEME\Dataset\Raw\Multimodality_understanding_text\d_b.csv")

df = pd.concat([d_a, d_b], ignore_index=True)
print(df.head())
```

```python
df = df[['image_file', 'Visual Elements']]
print(df.head())
```

```python
def extract_keywords_dict(df, nlp):
    """
    Returns a dictionary mapping image_file to a list of unique keywords extracted from Visual Elements.
    """
    def extract_keywords(text):
        doc = nlp(text)
        keywords = set()
        for chunk in doc.noun_chunks:
            keywords.add(chunk.root.text)
        for token in doc:
            if token.pos_ == "NOUN":
                keywords.add(token.text)
        return list(keywords)
    
    return {
        row['image_file']: extract_keywords(row['Visual Elements'])
        for _, row in df.iterrows()
    }

image_keywords = extract_keywords_dict(df, nlp)
print(image_keywords)
```

```python
# Save the image_keywords dictionary to a CSV file
image_keywords_df = pd.DataFrame([
    {'image_file': k, 'keywords': ', '.join(v)}
    for k, v in image_keywords.items()
])
image_keywords_df.to_csv(r"F:\Github\MemePic\MemePic\Met MEME\Dataset\Raw\Multimodality_understanding_text\image_keywords.csv", index=False)
```
