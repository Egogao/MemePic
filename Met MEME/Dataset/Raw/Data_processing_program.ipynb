```python
import os
import pandas as pd
from PIL import Image
import time
import logging
from tenacity import retry, stop_after_attempt, wait_fixed
from gradio_client import Client

# 初始化日志配置
logging.basicConfig(
    filename='processing_errors.log',
    level=logging.ERROR,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

# 初始化客户端
client = Client("http://127.0.0.1:7860/")

# 结果保存路径
OUTPUT_CSV = 'C:/Users/14122/Desktop/test data/Met-Meme/test_re_add.csv'

# 定义重试装饰器（3次重试，每次间隔1秒）
@retry(stop=stop_after_attempt(3), wait=wait_fixed(1))
def safe_predict(image_path):
#    prompt = """
#    Explain this meme. Follow ​**strictly** this format:
#    <IMAGE_ANALYSIS>
#    Visual Elements : [Descripted content]
#    Symbolic Significance : [Descripted content]
#    Overall Tone : [Descripted content]
#    </IMAGE_ASNALYSIS>
#
#    - Use exactly the same section titles and symbols (:)
#    """
    PROMPT_TEMPLATE = (
        r"Explain this meme. Follow  ​**strictly** this format:"
        r"\n"  # 显式换行符控制
        r"\n<IMAGE_ANALYSIS>"
        r"\nVisual Elements : [Descripted content]"
        r"\nSymbolic Significance : [Descripted content]"
        r"\nOverall Tone : [Descripted content]"
        r"\n</IMAGE_ANALYSIS>"
        r"\n\n- Use exactly the same section titles and symbols (:)"
    )

    pmpt = """Explain this meme. Follow **strictly** this format:
    <IMAGE_ANALYSIS>
    Visual Elements : [Descripted content]
    Symbolic Significance : [Descripted content]
    Overall Tone : [Descripted content]
    </IMAGE_ANALYSIS>
    
    - Use exactly the same section titles and symbols (:)"""

    pt = """Explain this meme. Follow **strictly** this format: <IMAGE_ANALYSIS> Visual Elements : [Descripted content]; Symbolic Significance : [Descripted content]; Overall Tone : [Descripted content] </IMAGE_ANALYSIS>
    
    - Please strictly use exactly the same section titles and symbols (:)"""       

    return client.predict(
        image_path,
        #"Explain this meme, Please use the following format the express the results: <IMAGE_ANALYSIS>; Visual Elements :: 'your description output'; Symbolic Significance :: 'your description output'; Overall Tone :: 'your description output'; <IMAGE_ANALYSIS>",  #question
        #prompt,
        pt,
        42,    #seed
        #0.95,   #top_p
        0.9,   #top_p
        0.1,    #temperature
        fn_index=2    #api_name
    )

def is_image_valid(image_path):
    """验证图片是否损坏"""
    try:
        with Image.open(image_path) as img:
            img.verify()
        return True
    except (IOError, SyntaxError, Image.UnidentifiedImageError) as e:
        logging.error(f"损坏文件: {image_path} - {str(e)}")
        return False

def load_processed_files():
    """加载已处理成功的文件列表"""
    if os.path.exists(OUTPUT_CSV):
        df = pd.read_csv(OUTPUT_CSV)
        return set(df['image_file'].tolist())
    return set()

def save_progress(image_file, result):
    """增量保存结果到CSV"""
    df = pd.DataFrame([{'image_file': image_file, 'result': result}])
    df.to_csv(OUTPUT_CSV, mode='a', header=not os.path.exists(OUTPUT_CSV), index=False, encoding='utf-8')

# 主处理逻辑
def process_images(image_directory):
    processed_files = load_processed_files()
    
    for image_file in os.listdir(image_directory):
        if image_file in processed_files:
            continue
            
        image_path = os.path.join(image_directory, image_file)
        
        # 1. 验证图片完整性
        if not is_image_valid(image_path):
            continue
        
        # 2. 处理并重试
        try:
            result = safe_predict(image_path)
            # 3. 验证响应格式
            if not isinstance(result, str) or "<IMAGE_ANALYSIS>" not in result:
                raise ValueError("无效的响应格式")
                
            # 4. 保存结果
            save_progress(image_file, result)
            print(f"处理成功: {image_file}")
            
        except Exception as e:
            logging.error(f"处理失败: {image_file} - {str(e)}", exc_info=True)
            print(f"处理失败: {image_file}，详见日志")

if __name__ == "__main__":
    #image_directory = image_directory  # 替换为实际路径
    #process_images(image_directory)
    process_images('C:/Users/14122/Desktop/test data/Met-Meme/filtered_images_add')
    print(f"处理完成，结果已保存至 {OUTPUT_CSV}")
```

```python
import re
import pandas as pd

def parse_image_analysis(df):
    # 改进后的正则表达式，支持 **Visual Elements** 等格式
    pattern = r"""
    <IMAGE_ANALYSIS>\s*  # 匹配 <IMAGE_ANALYSIS> 标签，后面可以有零个或多个空白字符
    (?:\*\*)?Visual\s*Elements(?:\*\*)?:?\s*(.*?)\s*  # 匹配 Visual Elements，支持可选的 ** 和 :
    (?:\*\*)?Symbolic\s*Significance(?:\*\*)?:?\s*(.*?)\s*  # 匹配 Symbolic Significance，支持可选的 ** 和 :
    (?:\*\*)?Overall\s*Tone(?:\*\*)?:?\s*(.*?)\s*  # 匹配 Overall Tone，支持可选的 ** 和 :
    </IMAGE_ANALYSIS>  # 匹配 </IMAGE_ANALYSIS> 标签
    """
    
    # 使用正则表达式提取内容
    extracted = df['result'].str.extract(
        re.compile(pattern, flags=re.DOTALL | re.IGNORECASE | re.VERBOSE)
    )
    
    # 重命名提取的列
    extracted.columns = ['Visual Elements', 'Symbolic Significance', 'Overall Tone']
    
    # 去除提取内容的前后空白字符
    extracted = extracted.apply(lambda x: x.str.strip() if x is not None else x)
    
    # 将提取的内容合并到原 DataFrame 中
    return pd.concat([df, extracted], axis=1)

# 假设 new_df 是你的原始 DataFrame
new_df_cleaned = parse_image_analysis(new_df)

# 查看提取后的结果
print(new_df_cleaned.head())
```
